```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense, Activation
import os
import random
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Model
from PIL import Image
```

```python
mobilenet= tf.keras.applications.mobilenet.MobileNet(include_top=False, weights='imagenet', input_shape=(224,224,3))
```

```python
angledata= pd.read_csv('/kaggle/input/carimages/Images/data.txt', names=['image_name', 'steering_angle'], delimiter=' ')
```

```python
x= mobilenet.output
x= tf.keras.layers.MaxPooling2D()(x)
x= tf.keras.layers.Dense(1024, activation= 'relu')(x)
x= tf.keras.layers.Dense(512, activation= 'relu')(x)
x= tf.keras.layers.Dropout(0.7)(x)
x= tf.keras.layers.Dense(512, activation= 'relu')(x)
x= tf.keras.layers.Dense(256, activation= 'relu')(x)
output= tf.keras.layers.Dense(1)(x)
model= tf.keras.Model(inputs= mobilenet.input, outputs= output)
```

```python
images=[]
import csv
import cv2
path= '/kaggle/input/carimages/Images/images/'
with open('/kaggle/input/carimages/Images/data.txt') as f:
    reader = csv.reader(f)
    for i,row in enumerate(reader):
        if i>2499:
            break
        x=path+row[0]
        x=x.split()
        img_path= str(x[0])
        
        img = cv2.imread(img_path)
        image= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        images.append(image)
```

```python
images_processed=[]
for i in images:
    single_img= Image.fromarray(i)
    single_img= single_img.resize((224, 224))
    single_img= np.reshape(single_img,(224, 224, 3))
    single_img= single_img/255.0
    images_processed.append(single_img)
```

```python
angledata=angledata['steering_angle']
labels= angledata.tolist()
```

```python
y_train= labels[:2200]
y_valid= labels[2200: 2500]

X_train= images_processed[:2200]
X_valid= images_processed[2200:]
```

```python
print('y_train:',len(y_train))
print('y_valid:',len(y_valid))
print('X_train:',len(X_train))
print('X_valid:',len(X_valid))

y_train= np.array(y_train)
y_valid= np.array(y_valid)

X_train= np.array(X_train)
X_valid= np.array(X_valid)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5), loss='mean_absolute_error')

model.fit(X_train,
    y_train,
    batch_size=128,
    epochs=100,
    validation_data=(X_valid,y_valid),
   )
```
